```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Iris.csv")
```

```python
df[:100]
```

```python
import matplotlib.pyplot as plt

x_sepal = df.SepalLengthCm
y_petal = df.PetalLengthCm

color_sepal = 'blue'
color_petal = 'green'

plt.scatter(x_sepal, y_petal, c=[color_sepal]*len(x_sepal), edgecolors='red', label='Sepal and Petal')

plt.xlabel('Length_sepal')
plt.ylabel('Petal_length')
plt.title('Scatter Plot of Sepal and Petal Dimensions')
plt.legend()
plt.show()

```
